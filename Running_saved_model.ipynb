```python
import pytesseract
from PyPDF2 import PdfReader
import csv

# Function to extract text from PDF using OCR
def extract_text_from_pdf(pdf_path):
    pdf = PdfReader(pdf_path)
    text = ""
    for page in pdf.pages:
        text += page.extract_text()
    return text

# Function to convert text to CSV format
def convert_text_to_csv(text, csv_path):
    lines = text.split("\n")
    with open(csv_path, 'w', newline='') as csv_file:
        csv_writer = csv.writer(csv_file)
        for line in lines:
            csv_writer.writerow([line])

# Convert PDF to CSV using OCR
def pdf_to_csv_with_ocr(pdf_path, csv_path):
    text = extract_text_from_pdf(pdf_path)
    convert_text_to_csv(text, csv_path)

# Example usage for PDF to CSV conversion
pdf_path = 'Apple Energy LLC_03.2023_Invoice 5828.pdf'
csv_path = 'Apple Energy LLC_03.2023_Invoice 5828.csv'
pdf_to_csv_with_ocr(pdf_path, csv_path)
```

```python
import pytesseract
from PyPDF2 import PdfReader
import csv
# Function to extract text from PDF using OCR
def extract_text_from_pdf(pdf_path):
    pdf = PdfReader(pdf_path)
    text = ""
    for page in pdf.pages:
        text += page.extract_text()
    return text
# Function to convert text to CSV format
def convert_text_to_csv(text, csv_path):
    lines = text.split("\n")
    with open(csv_path, 'w', newline='') as csv_file:
        csv_writer = csv.writer(csv_file)
        for line in lines:
            if line.strip():
                csv_writer.writerow([line.strip()])
# Convert PDF to CSV using OCR
def pdf_to_csv_with_ocr(pdf_path, csv_path):
    text = extract_text_from_pdf(pdf_path)
    convert_text_to_csv(text, csv_path)
# Example usage for PDF to CSV conversion
pdf_path = 'TPS-INVOICE-CALFLATG-Apr-2023.pdf'
csv_path = 'TPS-INVOICE-CALFLATG-Apr-2023_test.csv'
pdf_to_csv_with_ocr(pdf_path, csv_path)
```

```python
import csv

# Specify the path to your CSV file
csv_path = 'Apple Energy LLC_03.2023_Invoice 5828.csv'

# Specify the path and name of the output text file
text_path = 'output_calflag2.txt'

# Open the CSV file for reading
with open(csv_path, 'r') as csv_file:
    csv_reader = csv.reader(csv_file)

    # Open the text file for writing
    with open(text_path, 'w') as text_file:
        # Iterate over each row in the CSV file
        for row in csv_reader:
            # Join the values of the row with a delimiter (e.g., tab or space)
            line = '\t'.join(row)  # Use '\t' for tab-delimited or ' ' for space-delimited

            # Write the line to the text file
            text_file.write(line + '\n')
```
